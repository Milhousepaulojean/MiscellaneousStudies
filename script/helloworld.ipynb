```python
import os
import subprocess

# Descobre o JAVA_HOME automaticamente
result = subprocess.run(
    "dirname $(dirname $(readlink -f $(which java)))",
    shell=True, capture_output=True, text=True
)
java_home = result.stdout.strip()

# Força as variáveis para o PySpark
os.environ["JAVA_HOME"]    = java_home
os.environ["PATH"]         = java_home + "/bin:" + os.environ["PATH"]
os.environ["SPARK_LOCAL_IP"] = "127.0.0.1"  # resolve o aviso do hostname também

print("JAVA_HOME configurado:", java_home)

# Agora inicia o Spark
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("teste") \
    .master("local[*]") \
    .config("spark.driver.host", "127.0.0.1") \
    .getOrCreate()

spark.sparkContext.setLogLevel("ERROR")

dados = [("Paulo", 1), ("Jupyter", 2), ("PySpark", 3)]
df = spark.createDataFrame(dados, ["nome", "id"])
df.show()

print("Spark versão:", spark.version)
```
